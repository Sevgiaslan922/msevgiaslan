```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,],[7,3,9])
```

```python
x=np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x))
```

```python
plt.plot(x,np.sin(x), label= 'sin(x)')
plt.plot(x,np.cos(x), label= 'cos(x)')
plt.legend()
```

```python
plt.plot(x,np.sin(x), label= 'sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend(loc='upper center') #upper left upper right#or best

```

```python
plt.plot(x, np.sin(x), x, np.cos(x), x, x**2)
```

```python
plt.plot(x, np.sin(x), x, np.cos(x), x, x**2)
plt.xlabel ("x degerleri")
plt.ylabel ("y degerleri")
plt.title("x, y grafigi")
```
